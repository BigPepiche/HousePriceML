```python
import scripts.stats as st
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from sklearn import datasets
from importlib import reload

reload(st)
```

```python
californiaHousing = datasets.fetch_california_housing(data_home="./data/", download_if_missing=True)

df = pd.DataFrame(californiaHousing.data, columns=californiaHousing.feature_names)
df["price"] = californiaHousing.target
df.head(-1)
```

```python
StatList = ["Median", "Average", "Max", "Min", "Mode","Range", "Standard Deviation", "Variance", "IQR", "CV"]
sampleLength    = df.index[-1]
columnNames = californiaHousing.feature_names
stats = pd.DataFrame(None, columns=columnNames,index=StatList)
stats.head(-1)
```

```python
# Calcul de la médianne
for i in range(len(columnNames)):
    med = st.Stats.findMedian(df, columnNames[i])
    stats.loc["Median", columnNames[i]] = med
stats.head()
```

```python
# Calcul de la moyenne
for i in range(len(columnNames)):
    avg = st.Stats.findAverage(df, str(columnNames[i]))
    stats.loc["Average", columnNames[i]] = avg
stats.head()
```

```python
# Trouver le Max
for i in range(len(columnNames)):
    max = st.Stats.findMax(df, str(columnNames[i]))
    stats.loc["Max", columnNames[i]] = max
stats.head()
```

```python
# Trouver le min
for i in range(len(columnNames)):
    min = st.Stats.findMin(df, str(columnNames[i]))
    stats.loc["Min", columnNames[i]] = min
stats.head()
```

```python
# Touver le mode
for i in range(len(columnNames)):
    mode = st.Stats.findMode(df, str(columnNames[i]))
    stats.loc["Mode", columnNames[i]] = mode
stats.head()
```
