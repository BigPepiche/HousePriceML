```python
californiaHousing = datasets.fetch_california_housing(data_home="./data/", download_if_missing=True)

df = pd.DataFrame(californiaHousing.data, columns=californiaHousing.feature_names)
df["price"] = californiaHousing.target
df.head(-1)
```

```python
statFuncs = {
    "Median":st.Stats.median, 
    "Average":st.Stats.average,
    "Max":st.Stats.max, 
    "Min":st.Stats.min, 
    "Mode":st.Stats.mode,
    "Range":st.Stats.range, 
    "Standard Deviation":st.Stats.standardDeviation, 
    "Variance":st.Stats.variance, 
    "IQR":st.Stats.iqr, 
    "CV":st.Stats.coeficientOfVariation
    }
sampleLength    = df.index[-1]
columnNames = californiaHousing.feature_names

stats = pd.DataFrame(None, columns=columnNames,index=statFuncs.keys())

for name, func in statFuncs.items():
    for col in columnNames:
        stats.loc[name,col] = func(df, col)


stats.head(-1)
```
